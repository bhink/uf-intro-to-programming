## Why, What and How to Automate with Python

### Why Automate?

- Speed up data processing and data mining

- Autogenerated reporting (reports while you sleep!)

- More time for higher-level tasks and planning

### What to Automate?

| Easy to Automate                                                    | Difficult to Automate                     |
|---------------------------------------------------------------------|-------------------------------------------|
| Regularly scheduled reports or tasks                                | One-off or non-scheduled reporting        |
| Clear success / failure                                             | Unclear success / failure                 |
| All input can be handled via other machines or simple web interface | Requires deeper human input / instruction |
| If you fail, try try again                                          | If you fail, youâ€™re doomed                |

## How to Automate?

* When must this task begin?

* Does this task have a time limit? If so, when must it end?

* What are the necessary inputs for this task?

* What constitutes success, or partial success, for this task?

* If this task fails, what should happen?

* What does the task produce or provide? To whom? In what way?

* What, if anything, should happen after this task concludes?
